# WOE

```python
# Defining a function for calculating WOE and IV.
# For Discrete/Categorical.

def woe_cat(df, cat_feature, target):
    df = pd.concat([df[cat_feature], target], axis=1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WOE'] = np.log(df['prop_good'] / df['prop_bad'])
    df = df.sort_values(['WOE'])
    df = df.reset_index(drop=True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WOE'] = df['WOE'].diff().abs()
    df['IV'] = (df['prop_good'] - df['prop_bad']) * df['WOE']
    df['IV'] = df['IV'].sum()
    
    return df
```

```python
# Defining a function to plot WOE

def woePlot(df_woe, rotation_of_x_axis_labels=0):
    x = np.array(df_woe.iloc[:, 0].apply(str))
    y = df_woe['WOE']
    plt.figure(figsize=(14,8))
    plt.plot(x, y, marker='*', linestyle='--', color='b')
    plt.xlabel(df_woe.columns[0])
    plt.ylabel('Weight of Evidence (WOE)')
    plt.title(str('WOE by '+ df_woe.columns[0]))
    plt.xticks(rotation = rotation_of_x_axis_labels)
```

```python
# Defining a function for calculating WOE and IV.
# For Continuous variables.

def woe_num(df, cat_feature, target):
    df = pd.concat([df[cat_feature], target], axis=1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index=False)[df.columns.values[1]].mean()], axis=1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WOE'] = np.log(df['prop_good'] / df['prop_bad'])
    #df = df.sort_values(['WOE'])
    #df = df.reset_index(drop=True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WOE'] = df['WOE'].diff().abs()
    df['IV'] = (df['prop_good'] - df['prop_bad']) * df['WOE']
    df['IV'] = df['IV'].sum()
    
    return df
```

```python
ref_cat = pd.DataFrame(ref_cat, columns=['Features'])
ref_cat['Coefficients'] = 0

ref_cat = ref_cat.drop([5,6], axis=0)
ref_cat
```

```python
# Merging Feature coefficients and reference tables.

Scorecard = pd.concat([summary, ref_cat])
Scorecard = Scorecard.reset_index()
Scorecard
```

```python
Scorecard['Base Features'] = Scorecard['Features'].str.split(':').str[0]
Scorecard
```

```python
min_score = 300
max_score = 850
```

```python
Scorecard.groupby('Base Features')['Coefficients'].min()
```

```python
min_sum_coef = Scorecard.groupby('Base Features')['Coefficients'].min().sum()
min_sum_coef
```

```python
Scorecard.groupby('Base Features')['Coefficients'].max()
```

```python
max_sum_coef = Scorecard.groupby('Base Features')['Coefficients'].max().sum()
max_sum_coef
```

```python
Scorecard['Score'] = Scorecard['Coefficients'] * (max_score - min_score)/(max_sum_coef - min_sum_coef)
Scorecard
```

```python
score_min = Scorecard.groupby('Base Features')['Score'].min().sum()
score_min
```

```python
score_max = Scorecard.groupby('Base Features')['Score'].max().sum()
score_max
```

```python
# Scaling the scorecard between 300 and 850.

Scorecard['Score'][0] = ((Scorecard['Coefficients'][0] - min_sum_coef)/(max_sum_coef - min_sum_coef)) * (max_score - min_score) + min_score
Scorecard
```

```python
# Let's Confirm the maximum score.

maximum_score_sum = Scorecard.groupby('Base Features')['Score'].max().sum()
maximum_score_sum
```
